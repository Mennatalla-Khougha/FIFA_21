```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sb 
%matplotlib inline 
```

```python
df_full = pd.read_csv("players_21.csv")
df_full.head()
```

```python
df_full.info(verbose = True)
```

```python
df = df_full[["short_name", "age", "nationality", "club_name", "league_name", "league_rank", "overall", "potential", "value_eur", "wage_eur","player_positions", "preferred_foot", "work_rate"]]
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(axis = 0, inplace = True)
df.info()
```

```python
df["league_name"].unique()
```

```python
best = ["Italian Serie A", "Spain Primera Division", "German 1. Bundesliga", "French Ligue 1", "English Premier League"]
df = df.loc[df["league_name"].isin(best)]
df.info()
```

### What is/are the main feature(s) of interest in your dataset?

## Univariate Exploration

```python
plt.figure(figsize = [16,6])
base_color = sb.color_palette()[0]
plt.subplot(1, 2, 1)
bins = np.arange(59, df["potential"].max()+3, 2)
sb.countplot(data = df, x = "potential", order = bins, color = base_color);
plt.title("The top 5 league players potentials");
plt.subplot(1, 2, 2)
bins = np.arange(49, df["overall"].max()+3, 2)
sb.countplot(data = df, x = "overall", order = bins, color = base_color);
plt.title("The top 5 league players overall rating");
```

### The data show a gradual increase till 75 than a sharp decrease in the number of players with a higher rating and potential

```python
sb.countplot(data = df, x = "preferred_foot", color = base_color);
plt.title("The top 5 league players preferred food");
```

```python
df["work_rate"].unique()
```

```python
plt.figure(figsize = [10,6])
work = ["High/High", "High/Medium", "High/Low", "Medium/High", "Medium/Medium", "Medium/Low", "Low/High", "Low/Medium", "Low/Low"]
work = pd.api.types.CategoricalDtype(categories = work, ordered = True)
df["work_rate"] = df["work_rate"].astype(work)
sb.countplot(data = df, x = "work_rate", color = base_color);
plt.xticks(rotation = 15);
plt.title("The top 5 league players work rate");
```

### The majority of the player prefer their right foot as for work rate medium/medium is the highest

## Bivariate Exploration

```python
plt.figure(figsize = [18, 8])
plt.subplot(1, 2, 1)
plt.scatter(data = df, x = "overall", y = "age",alpha = 1/5);
plt.ylabel('players age')
plt.xlabel('players overall rating');
plt.title("The relation beteen age and overall rating")
plt.subplot(1, 2, 2)
plt.scatter(data = df, x = "potential", y = "age",alpha = 1/5);
plt.ylabel('players age')
plt.xlabel('players potential rating');
plt.title("The relation beteen age and thier potentials");
```

```python
plt.figure(figsize = [18, 6])
plt.subplot(1, 2, 1)
plt.hist2d(data = df, x = 'overall', y = 'age', cmin=0.5, cmap='viridis_r')
plt.colorbar()
plt.ylabel('players age')
plt.xlabel('players overall rating');
plt.title("The relation beteen age and overall rating")
plt.subplot(1, 2, 2)
plt.hist2d(data = df, x = 'potential', y = 'age', cmin=0.5, cmap='viridis_r')
plt.colorbar()
plt.ylabel('players age')
plt.xlabel('players potential rating');
plt.title("The relation beteen age and thier potentials");
```

### The older the players get the less potential they have and the less overall rating they get.

```python
plt.figure(figsize = [18, 6])
plt.subplot(1, 2, 1)
plt.title("The top 5 league and their players overall rating");
sb.violinplot(data = df, x = 'league_name', y = 'overall', color = base_color, inner = "quartile")
plt.xticks(rotation = 15);
plt.subplot(1, 2, 2)
plt.title("The top 5 league and their players overall rating");
sb.boxplot(data = df, x = 'league_name', y = 'overall', color = base_color)
plt.xticks(rotation = 15);
```

```python
plt.figure(figsize = [18, 6])
plt.subplot(1, 2, 1)
sb.violinplot(data = df, x = 'league_name', y = 'potential', color = base_color, inner = "quartile")
plt.xticks(rotation = 15);
plt.title("The top 5 league and their players potentials")
plt.subplot(1, 2, 2)
sb.boxplot(data = df, x = 'league_name', y = 'potential', color = base_color)
plt.xticks(rotation = 15);
plt.title("The top 5 league and their players potentials");
```
